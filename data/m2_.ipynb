```python
df = pd.read_csv("C:/PythonWork/Final_Project/data/US_M2.csv", encoding='utf-8', thousands=',')


val = df['Value2']
vv = val.values
vv

US_M2 = pd.DataFrame(vv, index= df['Date'], columns=['Value2'])
US_M2

# US_M2['Value'].plot()
```
